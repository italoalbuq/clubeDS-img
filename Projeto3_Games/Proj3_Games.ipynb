# 1.4 Tratamento de dados faltantes

```python
# Preencher um dado faltante da coluna Publisher
df1['Publisher'] = df1['Publisher'].fillna("Infogrames")

# De forma alternativa
#value1 = {"Publisher": 'Infogrames'}
#df1['Publisher'] = df1.fillna(value=value1)
```

```python
# Mediana de vendas do continente Norte Americano
df1["NorthAmerica_Sales"].median()
```

```python
# Preencher os dados faltantes de todas as colunas com o comando lambda
# Colocar a mediana de vendas de todos os continentes

# Para todos os valores x: 
# Os valores que são não-numéricos NaN (quando é retornado True), atribuir o valor da mediana
# Os valores que são númericos (quando é retornado False), atribuir/manter o mesmo valor numérico

df1['NorthAmerica_Sales'] = df1['NorthAmerica_Sales'].apply( lambda x: 3.28 if math.isnan( x ) else x )

df1['EurpeanUnion_Sales'] = df1['EurpeanUnion_Sales'].apply(lambda x: 2.26 if math.isnan(x) else x)

df1['Japan_Sales'] = df1['Japan_Sales'].apply( lambda x: 0.65 if math.isnan(x) else x)

df1['Other_Sales'] = df1['Other_Sales'].apply( lambda x: 0.63 if math.isnan(x) else x)

df1['Global_Sales'] = df1['Global_Sales'].apply( lambda x: 7.13 if math.isnan(x) else x)
```
